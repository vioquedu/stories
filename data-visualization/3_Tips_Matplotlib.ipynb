# 3 Tips to improve your charts with Matplotlib

Below the code used to write [this Medium post]()

## Import libraries

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
y = np.asarray([30, 45, 49, 76])
x = np.arange(4)
labels = ["ABC", "DEF", "UVW", "XYZ"]
```

## Default chart

```python
_, ax = plt.subplots(figsize=(10, 8))

ax.barh(x, y)

ax.set_yticks(x)
ax.set_yticklabels(labels, fontsize=16)

ax.set_title("Annual revenue of main players", fontsize=24)

plt.show()
```

## Decluttering

```python
_, ax = plt.subplots(figsize=(10, 8))

ax.barh(x, y)

ax.set_xticks([])
ax.set_yticks(x)
ax.set_yticklabels(labels, fontsize=16)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

for i in x:
    ax.text(y[i]-0.50, i, "{:0.0f}".format(y[i]), ha="right", color="white", fontsize=18, va="center")

ax.set_title("Annual revenue of main players", fontsize=24)

plt.show()
```

## Use color wisely

```python
_, ax = plt.subplots(figsize=(10, 8))


ax.barh(x[0:2], y[0:2], color="darkgray")
ax.barh(x[2], y[2], color="cornflowerblue", alpha=0.6)
ax.barh(x[3], y[3], color="royalblue")

ax.set_xticks([])
ax.set_yticks(x)
ax.set_yticklabels(labels, fontsize=16)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)


for i in x:
    weight = "bold" if i>1 else None
    ax.text(y[i]-0.50, i, "{:0.0f}".format(y[i]), ha="right", color="white", fontsize=18, va="center", weight=weight)
    
ax.set_title("Annual revenue of main players", fontsize=24, color="gray")

plt.show()
```

## Add annotations to emphasize

```python
_, ax = plt.subplots(figsize=(10, 8))

y1 = y[::-1]

ax.bar(x[2:], y1[2:], color="darkgray")
ax.bar(x[1], y1[1], color="cornflowerblue", alpha=0.6)
ax.bar(x[0], y1[0], color="royalblue")

ax.set_yticks([])
ax.set_xticks(x)
ax.set_xticklabels(labels[::-1], fontsize=16)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

for i in x:
    weight = "bold" if i<=1 else None
    ax.text(i, y1[i]-0.5, "{:0.0f}".format(y1[i]), va="top", color="white", fontsize=18, ha="center", weight=weight)

diff = (y1[0] - y1[1])
text = """XYZ leads the market
with a difference of
~{:0.0f} Million over
the second player""".format(diff)
ax.text(0.5, y1[0], text, ha="left", color="royalblue", fontsize=14, va="top")


ax.set_title("Annual revenue of main players", fontsize=24, color="gray")

plt.show()
```

## Bonus! Add Latex formatting to text

```python
_, ax = plt.subplots(figsize=(10, 8))

y1 = y[::-1]

ax.bar(x[2:], y1[2:], color="darkgray")
ax.bar(x[1], y1[1], color="cornflowerblue", alpha=0.6)
ax.bar(x[0], y1[0], color="royalblue")

ax.set_yticks([])
ax.set_xticks(x)
ax.set_xticklabels(labels[::-1], fontsize=16)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

for i in x:
    weight = "bold" if i<=1 else None
    ax.text(i, y1[i]-0.5, "{:0.0f}".format(y1[i]), va="top", color="white", fontsize=18, ha="center", weight=weight)

    
diff = (y1[0] - y1[1])
bold_text = r"$ \bf{ \sim" + f"{int(diff)}" + "\ Million}$"
text = f"""XYZ leads the market
with a difference of
{bold_text} over
the second player"""
ax.text(0.5, y1[0], text, ha="left", color="royalblue", fontsize=14, va="top")
    
ax.set_title("Annual revenue of main players", fontsize=24, color="gray")  
plt.show()
```
